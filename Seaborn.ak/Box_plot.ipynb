# Box Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var=sns.load_dataset("tips")
var
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var,hue="time")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var,hue="sex",color='g')
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var,order=["Fri","Sun","Thur","Sat"],
            showmeans=True,meanprops={"marker":"+","markeredgecolor":"b"},linewidth=2,
           palette="plasma",orient='v')
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(y="day",x="total_bill",data=var,order=["Fri","Sun","Thur","Sat"],
            showmeans=True,meanprops={"marker":"+","markeredgecolor":"b"},linewidth=2,
           palette="plasma",orient='h')
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(data=var,
            showmeans=True,meanprops={"marker":"+","markeredgecolor":"b"},linewidth=2,
           palette="plasma",orient='h')
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(y=var["total_bill"])
plt.show()
```
