# Violine Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var=sns.load_dataset("tips")
var
```

```python
sns.violinplot(x="day",y="total_bill",data = var,hue="time")
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data = var,linewidth=3)
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data = var,palette="Dark2")
plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",data = var,order=["Dinner","Lunch",])
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data = var,palette="Dark2",saturation=0.4)
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data = var,color='r')
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data = var,hue="sex",split=True,scale="count")
plt.show()
```

```python
sns.violinplot(y="day",x="total_bill",data = var,hue="sex")
plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",data = var,order=["Dinner","Lunch",],inner="quart")
plt.show()
```

```python
sns.violinplot(y=var["total_bill"])
plt.show()
```
