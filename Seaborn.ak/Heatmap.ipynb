# HeatMap

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
var= np.linspace(1,10,20).reshape(4,5)
var
```

```python
sns.heatmap(var)
```

```python
ak=sns.load_dataset("anagrams")
x=ak.drop(columns=["attnr"],axis=1).head(10)
x
```

# cmap= change the color the heatmap

```python
sns.heatmap(x,vmin=0,vmax=12,cmap="PuOr")
```

# annot to show the values in heatmap

```python
sns.heatmap(x,vmin=0,vmax=12,cmap="PuOr",annot=True)
```

```python
var_1=np.linspace(1,10,10).reshape(2,5)
var_1
```

```python
sns.heatmap(var_1,vmin=0,vmax=12,cmap="PuOr",annot=True)
```

```python
ar=np.array([["a0","a1","a2","a3","a4"],
             ["b0","b1","b2","b3","b4",]])
ar
```

# annot,annot_kws,

```python
y={"fontsize":20,"color":"r"}
sns.heatmap(var_1,vmin=1,vmax=10,cmap="PuOr",annot=ar,fmt="s",annot_kws=y)
```

# linewidth increase width b/w boxes

```python
y={"fontsize":20,"color":"r"}
sns.heatmap(var_1,vmin=1,vmax=10,cmap="PuOr",annot=ar,fmt="s",annot_kws=y,linewidth=4,linecolor='y')
```

# cbar => remove color bar
# xticklabels,yticklabels => remove labels

```python
y={"fontsize":20,"color":"r"}
sns.heatmap(var_1,vmin=1,vmax=10,cmap="PuOr",annot=ar,fmt="s",annot_kws=y,linewidth=4,
            linecolor='y',cbar=False,xticklabels=False,yticklabels=False)
plt.show()
```

# self made labels

```python
y={"fontsize":20,"color":"r"}
v=sns.heatmap(var_1,vmin=1,vmax=10,cmap="PuOr",annot=ar,fmt="s",annot_kws=y,linewidth=4,
            linecolor='y',cbar=False,xticklabels=False,yticklabels=False)
v.set(xlabel="Python",ylabel="Akash")
sns.set(font_scale=5)
plt.show()
```
