# Strip Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
var= sns.load_dataset("tips")
var
```

```python
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="rocket_r",linewidth=1,
              edgecolor="r",jitter=0.4)
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="rocket_r",linewidth=1,
              edgecolor="r",size=5,marker="<")
plt.show()
```

```python
m={"Male":"<","Female":"o"}
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="rocket_r",linewidth=1,
              edgecolor="r",size=5,marker="<",alpha=0.5)
plt.show()
```
